```python
# Install required packages if not already installed
!pip install faiss-cpu sentence-transformers langchain-huggingface PyPDF2 python-docx

# Import required libraries
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
import faiss
import PyPDF2
import docx
import os
from dotenv import load_dotenv

# Load environment variables for Hugging Face API key (if needed for HuggingFaceEndpoint)
load_dotenv()
print("API Token Set:", bool(os.environ.get('HUGGINGFACEHUB_API_TOKEN')))

# Define HuggingFaceEmbeddings (from previous context)
model_name = "sentence-transformers/all-MiniLM-L6-v2"
model_kwargs = {'device': 'mps'}  # Use 'cpu' if MPS is unavailable
encode_kwargs = {'normalize_embeddings': False}
hf_embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs
)

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_path):
    try:
        with open(pdf_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = ""
            for page in reader.pages:
                extracted = page.extract_text()
                if extracted:
                    text += extracted + "\n"
            return text.strip()
    except Exception as e:
        print(f"Error reading {pdf_path}: {e}")
        return ""

# Function to extract text from a Word document
def extract_text_from_docx(docx_path):
    try:
        doc = docx.Document(docx_path)
        text = ""
        for para in doc.paragraphs:
            text += para.text + "\n"
        return text.strip()
    except Exception as e:
        print(f"Error reading {docx_path}: {e}")
        return ""

# Define the texts to add to the vector store
texts = []
document_dir = '.'  # Replace with your document directory, e.g., '/path/to/documents'
for file in os.listdir(document_dir):
    if file.endswith('.pdf'):
        text = extract_text_from_pdf(os.path.join(document_dir, file))
        if text:
            texts.append(text)
    elif file.endswith('.docx'):
        text = extract_text_from_docx(os.path.join(document_dir, file))
        if text:
            texts.append(text)

# If no documents are found, use sample texts for testing
if not texts:
    print("No documents found. Using sample texts.")
    texts = ["Sample document for FAISS vector store.", "Example text for embedding."]

# Initialize FAISS index (384 dimensions for all-MiniLM-L6-v2)
index = faiss.IndexFlatL2(384)  # Replace with your custom index if defined elsewhere

# Create FAISS vector store with the embedding function
vector_store = FAISS(
    embedding_function=hf_embeddings.embed_query,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={}
)

# Add documents to the vector store
vector_store.add_texts(texts)

# Test the vector store with a sample query
query = "What is in the documents?"
results = vector_store.similarity_search(query, k=2)
for i, result in enumerate(results):
    print(f"Result {i+1}: {result.page_content[:100]}...")

def answer_question(vectorstore, query):
    """Answers a question based on the provided vectorstore."""
    llm = HuggingFaceEndpoint(repo_id= 'meta-llama/Meta-Llama-3-8B-Instruct', 
                              token = huggingface_api_key, temperature= 0.6)
    qa = RetrievalQA.from_chain_type(llm=llm, retriever=vectorstore.as_retriever())

    answer = qa({"query": query})
    return answer
```
